```python
X=df.drop('target',axis='columns')
y=df['target']
```

```python
from sklearn.model_selection import train_test_split as tts
X_train,X_test,y_train,y_test=tts(X,y, test_size=0.2)
```

```python
from sklearn.svm import SVC
model=SVC(kernel='linear')
model.fit(X_train,y_train)
```

```python
y_pre=model.score(X_test,y_test)
print(f'Support vector machine model Accuracy :{y_pre*100:.2f}')
```

```python
from sklearn.ensemble import RandomForestClassifier as rfc
model=rfc()
model.fit(X_train,y_train)
```

```python
y_pre=model.score(X_test,y_test)
print(f"Random Forest Model Accuracy : {y_pre*100:.2f}")
```

```python
from sklearn.tree import DecisionTreeClassifier as dtc
model=dtc()
model.fit(X_train,y_train)
```

```python
pre=model.score(X_test,y_test)
print(f"Decision Tree Model Accuracy : {pre*100:.2f}")
```
